```python
import pandas as pd
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler
data = {
    'name': ['Rex', 'Peter', 'Sam', 'Rita', 'Hilda'],
    'age': [1, 95, 22, 0, 60],
    'height': [20, 180, 160, 145, 125],
    'weight': [0, 100, 75, 55, 40],
    'bp': [80, 100, 65, 70, 40]
}

frame = DataFrame(data)
print(frame)
frame.to_csv('patient_data.csv', index=False)
df = pd.read_csv('patient_data.csv')
print(df)
df[['age', 'height', 'weight', 'bp']] = df[['age', 'height', 'weight', 'bp']].replace(0, df.mean(numeric_only=True))
print(df)
frame2 = df[['age', 'bp', 'height', 'weight']]
print(frame2)
scaler = StandardScaler()
scaled = scaler.fit_transform(frame2)

print(scaled)
print("Shape:", scaled.shape)
print("Mean:", scaled.mean(axis=0))
print("Standard deviation:", scaled.std(axis=0))
```
